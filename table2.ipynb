```python
import pandas as pd

# Odczytanie pliku CSV
df = pd.read_csv('t_test_results.csv')

# Lista unikalnych wartości 'crypto' i 'Classifier 2'
cryptos = df['crypto'].unique()
classifiers = df['Classifier 2'].unique()

# Generowanie kodu LaTeX
latex_code = """
\\documentclass{article}
\\usepackage[table]{xcolor}
\\begin{document}
\\scalebox{0.7}{
\\begin{tabular}{|l|""" + "c|"*len(classifiers) + """}
\\hline
 & """ + " & ".join(classifiers) + """ \\\\
\\hline
"""

# Iterowanie przez każdą kryptowalutę
for crypto in cryptos:
    latex_code += f"\\textbf{{{crypto}}} \\\\ \\hline\n"
    sub_df = df[df['crypto'] == crypto]
    classifiers_1 = sub_df['Classifier 1'].unique()
    
    for classifier_1 in classifiers_1:
        latex_code += classifier_1
        for classifier_2 in classifiers:
            if classifier_1 == classifier_2:
                latex_code += " & \\cellcolor{gray} "
            else:
                cell_df = sub_df[(sub_df['Classifier 1'] == classifier_1) & (sub_df['Classifier 2'] == classifier_2)]
                if not cell_df.empty:
                    t_statistic = cell_df['t-statistic'].values[0]
                    print(t_statistic)
                    p_value = cell_df['p-value'].values[0]
                    if t_statistic > 0 and p_value < 0.05:
                        latex_code += " & \\cellcolor{black} "
                    else:
                        latex_code += " & \\cellcolor{white} "
                else:
                    latex_code += " & \\cellcolor{white} "
        latex_code += " \\\\ \n"
    latex_code += "\\hline\n"

latex_code += """
\\end{tabular}}
\\end{document}
"""

# Zapisanie kodu LaTeX do pliku
with open('table.tex', 'w') as f:
    f.write(latex_code)

print("Kod LaTeX został wygenerowany i zapisany do pliku 'table.tex'.")
```
